```python
#------------------EDA Functions-------------------------

def openfile():    
    global file_path
    file_path = filedialog.askopenfilename(
            filetypes=(
            ("CSV files", "*.csv"),
            ("Excel files", "*.xlsx *.xls"),
            ("All files", "*.*")
        )
    )
openfile()
if file_path:
        # A file was selected, now determine the file type and read it
    try:
        if file_path.endswith('.csv'):
            df = pd.read_csv(file_path)
        elif file_path.endswith('.xlsx') or file_path.endswith('.xls'):
            df = pd.read_excel(file_path)
        else:
            print("Unsupported file type selected.")
        col_options=list(df.columns)
            
    except Exception as e:
        print(f"Error reading file: {e}")
        
else:
    show_box("No file selected.")
    


# -------------------------
# Text Info
# -------------------------    
def dfhead_func():
    show_box(df.head().to_string())
def dfshape_func():
    show_box(df.shape)
    
def dfdescribe_func():
    show_box(df.describe().to_string())
    
def dfisnull_func():
    show_box(df.isnull().sum().to_string())

def dfdropna_func():
    df.dropna(axis=0, inplace=True, ignore_index=True)
    show_box("Null Valued Rows dropped")

def dffillna_func(event=None):
    cb_var=tab1_combo_box.get()
    if cb_var=='Mean':
        df.fillna(df.mean(), inplace=True)
    elif cb_var=='Mode':
        df.fillna(df.mode().iloc[0], inplace=True)
    else:
        df.fillna(df.median(), inplace=True)
    
    show_box(f"Null Valued Filled Successflly with {cb_var} values")
    
def dfinfo_func():
    buffer = io.StringIO()
    df.info(buf=buffer)
    info_str = buffer.getvalue()
    show_box(info_str)

def show_box(text):
    text_box.configure(state="normal")
    text_box.delete("1.0", "end")
    text_box.insert("end", text)
    text_box.configure(state="disabled")

def correlation_func():
    global corr
    corr = df.corr(numeric_only=True)
    show_box(corr)
```

# ML Functions

```python
#-----------Machine Learning Functions ----------------------
def cls_textbox_ML():
    text_box_ML.configure(state="normal")
    text_box_ML.delete("1.0", "end")
    text_box_ML.configure(state="disabled")

def show_box_Lin_Reg(coef, intercept, y_pred,mae,mse,r2):
    message="Successfully Processed"
    text_box_ML.configure(state="normal")
    text_box_ML.delete("1.0", "end")
    text_box_ML.insert("end", message)
    text_box_ML.insert("end", "\nCoefficient   = ")
    text_box_ML.insert("end", coef)
    text_box_ML.insert("end", "\nIntercept     = ")
    text_box_ML.insert("end", intercept)
    text_box_ML.insert("end", "\nMean Absolute Error  =  ")
    text_box_ML.insert("end", mae)
    text_box_ML.insert("end", "\nMean Squared Error   =  ")
    text_box_ML.insert("end", mse)
    text_box_ML.insert("end", "\nR2 Score   =   ")
    text_box_ML.insert("end", r2)
    text_box_ML.configure(state="disabled")

def Linear_Reg_func():
    #global model, y_pred, coef, intercept,mae,mse,r2
    try:
        model = LinearRegression()
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        coef = model.coef_
        intercept = model.intercept_
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2  = r2_score(y_test, y_pred)
        show_box_Lin_Reg(coef, intercept,y_pred,mae,mse,r2)
    except:
        messagebox.showinfo("Error", "First Train the features")

def show_box_Log_Reg(coef, intercept,y_pred_log,y_proba_log,acc):
    message="Successfully Processed"
    text_box_ML.configure(state="normal")
    text_box_ML.delete("1.0", "end")
    text_box_ML.insert("end", message)
    text_box_ML.insert("end", "\nCoefficient   = ")
    text_box_ML.insert("end", coef)
    text_box_ML.insert("end", "\nIntercept     = ")
    text_box_ML.insert("end", intercept)
    text_box_ML.insert("end", "\ny_pred =\n")
    text_box_ML.insert("end", y_pred_log)
    text_box_ML.insert("end", "\ny_probability_log  =  ")
    text_box_ML.insert("end", y_proba_log)
    text_box_ML.insert("end", "\nAccuracy   =  ")
    text_box_ML.insert("end", acc)
    text_box_ML.configure(state="disabled")

def Logistic_Reg_func():
    try:
        log_model=LogisticRegression(max_iter=200)
        log_model.fit(X_train,y_train)
        y_pred_log = log_model.predict(X_test)          # the predict() method uses the default 0.5 threshold
        y_proba_log = log_model.predict_proba(X_test)
        coef = log_model.coef_
        intercept = log_model.intercept_
        acc = accuracy_score(y_test, y_pred_log)
        show_box_Log_Reg(coef, intercept,y_pred_log,y_proba_log,acc)
    except:
        messagebox.showinfo("Error", "First Train the features correctly")

```

## One-Hot Encoding

```python
def one_hot_encoder():
    global df_ohe
    encoding_select()
    encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    encoded_data = encoder.fit_transform(df[encoding_items])# .fit_transform expects a 2D array,
    # 2. Convert the resulting NumPy array back to a DataFrame with meaningful names
    ohe_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(encoding_items))
    # 3. Combine with original data (optional)
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    df_ohe= pd.concat([df.drop(categorical_columns, axis=1), ohe_df], axis=1)
    print(df_ohe)
```
